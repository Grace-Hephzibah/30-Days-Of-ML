## 3 sided polygon (Type 1)

### triangle(base, height)

![Right Angle Triangle](https://www.geogebra.org/resource/bjecpkag/G64k3AMviIMQyQRM/material-bjecpkag.png)

```python
def triangle (base, height):
    return 0.5*base*height
```

```python
triangle(5,10)
```

## 3 sided polygon (Type 2)

### triangle( side a, side b, side c)

```python
def triangle(a,b,c):
    s = (a+b+c)/2
    area = (s*(s-a)*(s-b)*(s-c))**0.5
    return area
```

```python
triangle(3,4,5)
```

## 4-sided polygon

```python
def square(s):
    return s*s
```

This __function__ takes in only 1 argument and returns the area of a square

The argument passed in is the value of a _side_

```python
def rectangle(a,b):
    return a*b
```

This __function__ takes in 2 arguments and returns the area of a rectangle

The argument passed in is the value of _both sides_

```python
square(5)
```

```python
rectangle(2,5)
```
